```python
# Import libraries
import pandas as pd # Dataframe (table) handling
import numpy as np # Numeric python, functions for calculations
import matplotlib.pyplot as plt # graphs
import plotly.express as px # more graphs
import plotly.graph_objects as go # and more graphs

# Directory path where the data is stored
path = 'https://raw.githubusercontent.com/prof-apartida/data-exercises/main/'
```

## Exploratory data analysis (EDA)

As mentioned, data exploration is crucial to determine if the information is accurate and has a minimum amount to get deeper insights.

### Import data

```python
df = pd.read_csv(path+'GDP_DATA.csv', header=2) # Import the data, use the data from the third row (the two forst rows have no important information, is noise)

# Function to remove the commas and convert them into numbers
def remove_commas(x):
    return x.replace(',', '') if isinstance(x, str) else x

df = df.applymap(remove_commas) # apply a function to the DataFrame
df = df.apply(pd.to_numeric, downcast='integer', errors='ignore') # downcast finds the minimum viable type for each column

df.head()# to see the first 5 rows
```

### Get some information about the dataframe (completeness, data types, column names)

```python
df.info()
```

We can see that in 1960, we have only 133 non-null values, which means that we have a completeness of 50% (266 rows). In 2014 we have 259 non-null values, which means that we have a completeness of 97.4% in that year

### Get some statistical information

```python
df.describe() # get a summary of the numerical columns
```

In 1960, the <b>minimum GDP</b> value was 1.201201e+07 (12012010) USD, while the <b>maximum</b> value was 1.392053e+12 (1392053000000) USD.

In 2014, the <b>minimum GDP</b> value was 3.875969e+07 (38759690) USD, while the <b>maximum</b> value was 7.970881e+13 (79708810000000) USD.

#### Graph to show outliers

This is to identify the countries with GDP values that differ significantly from the rest of the sample

```python
year = '2014'
df1 = df[['Country Code','Country Name',year]]

fig = px.box(df1, y=year, hover_data=['Country Name'])
fig.show()
```

### Get all possible values for column named "Country Name" and "Country code", so we can analyze if we should ignore rows (contries in this case)

```python
pd.set_option('display.max_rows', None) # ask pandas to show all rows
pd.set_option('display.max_columns', None) # ask pandas to show all columns

df[['Country Name','Country Code']]
```

```python
# Create a list excluding the non-valid "countries" (as they are not countries)
exclude = ['AFE', 'WLD','SAS','HIC','OED','PST','IBT','LMY','MIC','IBD','EAS','UMC','LTE','ECS','NAC','EAP','TEA','EUU','EMU','EAR','LMC','LCN','TLA','LAC','TEC','TSA','MEA','ECA','ARB','IDA','SSF','TSS','SSA','CEB','FCS','MNA','IDX','TMN','PRE','LDC','IDB','HPC','AFW','SST','LIC','OSS','CSS']

df_new = df[~df['Country Code'].isin(exclude)] # Create a dataframe that does not include the aforementioned codes
df_new.describe()

fig = px.box(df_new, y="1960", hover_data=['Country Name'])
fig.show()
```

## Data analysis
### Descriptive analytics: graphs
#### Bar graphs

```python
# Graph the GDP values by country of the year 2014
year = '2014'

fig = px.bar(df_new, x=year, y='Country Name', title="Countries' GDP")
fig.show()

fig = px.bar(df_new.sort_values(by=year, ascending=False).reset_index(drop=True), x=year, y='Country Name', title="Countries'  GDP") # Sort the table by ascending values
fig.show()

fig = px.bar(df_new.sort_values(by=year, ascending=False).head(15).reset_index(drop=True), x=year, y='Country Name', title='Top 15 countries by GDP')# Select the first 15 countries with higher GPD values
fig.show()

fig = px.bar(df_new.sort_values(by=year, ascending=False).tail(15).reset_index(drop=True), x=year, y='Country Name', title='Bottom 15 countries by GDP')# Select the first 15 countries with lower GPD values
fig.show()
```

#### Pie charts

World GDP contribution by country

```python
year = '2014'
fig = px.pie(df_new.sort_values(by=year, ascending=False).reset_index(drop=True), values=year, names='Country Name', title='Top countries by GDP in '+year)
fig.show()

# Pie chart 2
fig = px.pie(df_new.sort_values(by=year, ascending=False).head(5).reset_index(drop=True), values=year, names='Country Name', title='Pie chart 2: Top 5 countries by GDP in '+year)
fig.show()

# Pie chart 3
year = '1975'
fig = px.pie(df_new.sort_values(by=year, ascending=False).head(5).reset_index(drop=True), values=year, names='Country Name', title='Pie chart 3: Top 5 countries by GDP in '+year)
fig.show()
```

#### World map

```python
# Create a map to spatially illustrate the GDP values
year = '2014'

fig = px.choropleth(df_new, locations='Country Code', color=year, hover_data=['Country Name'], # hover is when you place the mouse over the map
                    color_continuous_scale='Reds',  # Define the color pallete
                    color_continuous_midpoint=np.mean(df_new[year]), # the midpoint color at the mean value
                    range_color=(np.min(df_new[year]),np.max(df_new[year])), # range of intensity from the minimum to the maximum valur
                    title="GDP in " + year + " by country" )
fig.show()


fig = px.choropleth(df_new, locations='Country Code', color=year, hover_data=['Country Name'], # hover is when you place the mouse over the map
                    color_continuous_scale='hot',  # Define the color pallete
                    color_continuous_midpoint=np.mean(df_new[year]), # the midpoint color at the mean value
                    range_color=(np.min(df_new[year]),np.max(df_new[year])), # range of intensity from the minimum to the maximum valur
                    title="GDP in " + year + " by country" )
fig.show()
```

```python
fig = px.colors.sequential.swatches_continuous()
fig.show()
```

<b> Table transposition: change the data from rows, to columns and viceversa </b><br>

We want to obtain data by country, and in the other table, the columns are by year, so we proceed to transpose the table.

```python
df_t = df_new.transpose() # transpose the table
df_t = df_t.reset_index() # reset the index
df_t.columns = df_t.iloc[1] # set the row with the country codes as column names, we select the country codes, so we have no problen on typing the name
df_t = df_t.drop(df_t.index[0:4]) # remove the first rows
df_t = df_t.rename(columns={'Country Code': 'Year'}) # Rename the column
df_t = df_t.reset_index(drop=True) # reset index and create a new one (drop=True)

df_t.head()
```

#### Line charts

```python
# Let's see the total GDP progress of Qatar
country = 'QAT'

fig = go.Figure()
fig.add_trace(go.Scatter(x=df_t['Year'], y=df_t[country], mode='lines', name=country))
fig.update_layout(title=country + " GDP progress")
fig.show()
```

```python
# Now, let's compare it with USA
country1 = 'QAT'
country2 = 'JPN'

fig = go.Figure()
fig.add_trace(go.Scatter(x=df_t['Year'], y=df_t[country1], mode='lines', name=country1))
fig.add_trace(go.Scatter(x=df_t['Year'], y=df_t[country2], mode='lines', name=country2))
fig.update_layout(title="GDP comparison")
fig.show()
```
