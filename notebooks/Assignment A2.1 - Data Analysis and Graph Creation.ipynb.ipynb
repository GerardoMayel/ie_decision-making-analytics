# Assignment A2.1 - Data Analysis and Graph Creation

This notebook creates the required visualizations for the GDP assignment.

Selected countries:

- Mexico (`MEX`)
- Spain (`ESP`)

Selected years:

- World ranking context year: `2014`
- Top 5 GDP pie chart year: `2018`

The notebook creates three graphs and one pie chart. It also includes the minimum value, maximum value, mean, and standard deviation for both selected countries.

## 1. Import libraries and define the data path

Requirement covered: prepare the data analysis code to load, clean, and visualize the GDP dataset.

```python
# Import libraries
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Data path used in the course notebook
path = 'https://raw.githubusercontent.com/prof-apartida/data-exercises/main/'
```

## 2. Load and clean the dataset

The GDP values are stored by country and year. Some values may include commas, so the code removes them before converting the year columns to numbers.

```python
# Load the GDP dataset
# The first two rows are not used, so header=2 starts the table in the correct row.
df = pd.read_csv(path + 'GDP_DATA.csv', header=2)

# Function to remove commas from values stored as text
def remove_commas(x):
    return x.replace(',', '') if isinstance(x, str) else x

# Remove commas from all columns
for col in df.columns:
    df[col] = df[col].map(remove_commas)

# Identify year columns and convert them to numeric values
year_cols = [col for col in df.columns if str(col).isdigit()]
for col in year_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Preview the cleaned dataset
df.head()
```

## 3. Keep only individual countries

Requirement covered: avoid comparing countries with regional, income, or world aggregate rows.

```python
# Codes that represent regions, income groups, or other aggregates
exclude = [
    'AFE', 'WLD', 'SAS', 'HIC', 'OED', 'PST', 'IBT', 'LMY', 'MIC', 'IBD',
    'EAS', 'UMC', 'LTE', 'ECS', 'NAC', 'EAP', 'TEA', 'EUU', 'EMU', 'EAR',
    'LMC', 'LCN', 'TLA', 'LAC', 'TEC', 'TSA', 'MEA', 'ECA', 'ARB', 'IDA',
    'SSF', 'TSS', 'SSA', 'CEB', 'FCS', 'MNA', 'IDX', 'TMN', 'PRE', 'LDC',
    'IDB', 'HPC', 'AFW', 'SST', 'LIC', 'OSS', 'CSS'
]

# Create a dataset with individual countries only
df_new = df[~df['Country Code'].isin(exclude)].copy()

print('Rows in the original dataset:', len(df))
print('Rows after removing aggregates:', len(df_new))
print('Number of year columns:', len(year_cols))
```

## 4. Select countries and years

Requirement covered: select two countries for the comparison, one year for the world position graph, and another valid year for the pie chart.

```python
# Selected countries for the assignment
country1 = 'MEX'
country2 = 'ESP'
selected_countries = [country1, country2]

# Year used to compare the selected countries with the rest of the world
ranking_year = '2014'

# Year used for the pie chart. It must not be 2014 or 1975.
pie_year = '2018'

if pie_year in ['2014', '1975']:
    raise ValueError('The pie chart year must be different from 2014 and 1975.')

# Get country names from the dataset
country_names = dict(zip(df_new['Country Code'], df_new['Country Name']))
missing_countries = [code for code in selected_countries if code not in country_names]
if missing_countries:
    raise ValueError(f'These country codes are not in the dataset: {missing_countries}')

country1_name = country_names[country1]
country2_name = country_names[country2]

print('Selected countries:', country1_name, 'and', country2_name)
print('World ranking year:', ranking_year)
print('Pie chart year:', pie_year)
```

## 5. Prepare the data in long format

This format makes it easier to create time series graphs and calculate descriptive statistics by country.

```python
# Convert the data from wide format to long format
# Wide format: one column for each year
# Long format: one row for each country and year

df_long = df_new.melt(
    id_vars=['Country Name', 'Country Code'],
    value_vars=year_cols,
    var_name='Year',
    value_name='GDP'
)

df_long['Year'] = df_long['Year'].astype(int)
df_long = df_long.dropna(subset=['GDP']).copy()

# Keep data for the two selected countries
selected_data = df_long[df_long['Country Code'].isin(selected_countries)].copy()

selected_data.head()
```

## 6. Helper function for readable GDP values

This function changes large GDP values into a shorter format, such as billions or trillions of US dollars.

```python
# Format GDP values for labels and tables
def format_gdp(value):
    if pd.isna(value):
        return ''
    if abs(value) >= 1_000_000_000_000:
        return f'${value / 1_000_000_000_000:.2f}T'
    if abs(value) >= 1_000_000_000:
        return f'${value / 1_000_000_000:.2f}B'
    if abs(value) >= 1_000_000:
        return f'${value / 1_000_000:.2f}M'
    return f'${value:,.0f}'
```

## 7. Calculate summary statistics for both countries

Requirement covered: minimum value, maximum value, mean, and standard deviation for both countries. The years for the minimum and maximum values are also calculated.

```python
# Calculate the required quantitative values for both countries
summary_rows = []

for code in selected_countries:
    temp = selected_data[selected_data['Country Code'] == code].sort_values('Year').copy()
    if temp.empty:
        raise ValueError(f'No GDP data found for {code}.')

    min_row = temp.loc[temp['GDP'].idxmin()]
    max_row = temp.loc[temp['GDP'].idxmax()]

    summary_rows.append({
        'Country Code': code,
        'Country Name': temp['Country Name'].iloc[0],
        'Minimum GDP': min_row['GDP'],
        'Minimum Year': int(min_row['Year']),
        'Maximum GDP': max_row['GDP'],
        'Maximum Year': int(max_row['Year']),
        'Mean GDP': temp['GDP'].mean(),
        'Standard Deviation': temp['GDP'].std()
    })

summary = pd.DataFrame(summary_rows)

# This table supports the second graph and helps verify the values.
summary_display = summary.copy()
for col in ['Minimum GDP', 'Maximum GDP', 'Mean GDP', 'Standard Deviation']:
    summary_display[col] = summary_display[col].apply(format_gdp)

summary_display
```

## Graph 1. GDP evolution over time

Requirement covered: plot the GDPs of the two selected countries on the same graph.

Mexico and Spain have GDP values in a similar range, so one shared y-axis is enough for comparison.

```python
# Requirement covered: first graph with both selected countries on the same chart
# A shared y-axis is used because Mexico and Spain have GDP values in a similar range.
country1_data = selected_data[selected_data['Country Code'] == country1].sort_values('Year')
country2_data = selected_data[selected_data['Country Code'] == country2].sort_values('Year')

fig1 = px.line(
    selected_data.sort_values('Year'),
    x='Year',
    y='GDP',
    color='Country Name',
    title=f'Graph 1. GDP Evolution Over Time: {country1_name} and {country2_name}',
    labels={
        'Year': 'Year',
        'GDP': 'GDP (current US$)',
        'Country Name': 'Country'
    },
    template='plotly_white'
)

fig1.update_traces(
    mode='lines',
    hovertemplate='Year: %{x}<br>GDP: $%{y:,.0f}<extra></extra>'
)
fig1.update_yaxes(tickprefix='$', tickformat='.2s')
fig1.update_layout(height=550, legend_title_text='Country')

fig1.show()
```

## Graph 2. GDP comparison with summary statistics

Requirement covered: include all required quantitative values for both countries.

This graph includes:

- minimum GDP value and year;
- maximum GDP value and year;
- mean GDP;
- standard deviation.

```python
# Requirement covered: second graph with all required statistics for both countries
# A shared y-axis keeps the comparison direct because both countries are in a similar GDP range.
fig2 = make_subplots(
    rows=2,
    cols=1,
    specs=[[{}], [{'type': 'table'}]],
    row_heights=[0.68, 0.32],
    vertical_spacing=0.08
)

# Add GDP lines
for code, data in [(country1, country1_data), (country2, country2_data)]:
    name = country_names[code]
    fig2.add_trace(
        go.Scatter(
            x=data['Year'],
            y=data['GDP'],
            mode='lines',
            name=f'{name} GDP',
            hovertemplate='Year: %{x}<br>GDP: $%{y:,.0f}<extra></extra>'
        ),
        row=1,
        col=1
    )

# Add mean lines and min/max markers
for code, data in [(country1, country1_data), (country2, country2_data)]:
    name = country_names[code]
    stats = summary[summary['Country Code'] == code].iloc[0]

    # Mean line
    fig2.add_trace(
        go.Scatter(
            x=[data['Year'].min(), data['Year'].max()],
            y=[stats['Mean GDP'], stats['Mean GDP']],
            mode='lines',
            line=dict(dash='dash'),
            name=f'{name} mean',
            hovertemplate='Mean GDP: $%{y:,.0f}<extra></extra>'
        ),
        row=1,
        col=1
    )

    # Minimum marker
    fig2.add_trace(
        go.Scatter(
            x=[stats['Minimum Year']],
            y=[stats['Minimum GDP']],
            mode='markers+text',
            name=f'{name} minimum',
            text=[f"Min {int(stats['Minimum Year'])}<br>{format_gdp(stats['Minimum GDP'])}"],
            textposition='bottom center',
            marker=dict(size=10),
            hovertemplate='Minimum GDP: $%{y:,.0f}<extra></extra>'
        ),
        row=1,
        col=1
    )

    # Maximum marker
    fig2.add_trace(
        go.Scatter(
            x=[stats['Maximum Year']],
            y=[stats['Maximum GDP']],
            mode='markers+text',
            name=f'{name} maximum',
            text=[f"Max {int(stats['Maximum Year'])}<br>{format_gdp(stats['Maximum GDP'])}"],
            textposition='top center',
            marker=dict(size=10),
            hovertemplate='Maximum GDP: $%{y:,.0f}<extra></extra>'
        ),
        row=1,
        col=1
    )

# Prepare the statistics table inside the same visual
summary_table = summary.copy()
summary_table['Minimum GDP and year'] = summary_table.apply(
    lambda row: f"{format_gdp(row['Minimum GDP'])}<br>{int(row['Minimum Year'])}", axis=1
)
summary_table['Maximum GDP and year'] = summary_table.apply(
    lambda row: f"{format_gdp(row['Maximum GDP'])}<br>{int(row['Maximum Year'])}", axis=1
)
summary_table['Mean GDP'] = summary_table['Mean GDP'].apply(format_gdp)
summary_table['Standard Deviation'] = summary_table['Standard Deviation'].apply(format_gdp)

fig2.add_trace(
    go.Table(
        header=dict(
            values=['Country', 'Minimum GDP<br>and year', 'Maximum GDP<br>and year', 'Mean GDP', 'Standard deviation'],
            align='center'
        ),
        cells=dict(
            values=[
                summary_table['Country Name'],
                summary_table['Minimum GDP and year'],
                summary_table['Maximum GDP and year'],
                summary_table['Mean GDP'],
                summary_table['Standard Deviation']
            ],
            align='center'
        )
    ),
    row=2,
    col=1
)

fig2.update_layout(
    title=f'Graph 2. GDP Comparison with Summary Statistics: {country1_name} and {country2_name}',
    legend_title_text='Country and statistic',
    template='plotly_white',
    height=850
)
fig2.update_xaxes(title_text='Year', row=1, col=1)
fig2.update_yaxes(title_text='GDP (current US$)', tickprefix='$', tickformat='.2s', row=1, col=1)

fig2.show()
```

## Graph 3. Position of the selected countries in the world ranking

Requirement covered: show the position of the selected countries in the context of the rest of the world for the selected year.

This graph uses all individual countries with available GDP data in the selected year. Rank 1 means the highest GDP.

```python
# Requirement covered: third graph with world context for the selected countries
ranking_df = df_new[['Country Name', 'Country Code', ranking_year]].dropna().copy()
ranking_df = ranking_df[ranking_df[ranking_year] > 0]
ranking_df = ranking_df.sort_values(by=ranking_year, ascending=False).reset_index(drop=True)
ranking_df['Rank'] = ranking_df.index + 1
ranking_df['Country group'] = np.where(
    ranking_df['Country Code'].isin(selected_countries),
    'Selected countries',
    'Other countries'
)

fig3 = px.scatter(
    ranking_df,
    x='Rank',
    y=ranking_year,
    color='Country group',
    hover_data={
        'Country Name': True,
        'Country Code': True,
        'Rank': True,
        ranking_year: ':,.0f',
        'Country group': False
    },
    log_y=True,
    title=f'Graph 3. World GDP Ranking in {ranking_year}: {country1_name} and {country2_name}',
    labels={
        'Rank': f'World rank in {ranking_year} (1 = highest GDP)',
        ranking_year: 'GDP (current US$, log scale)',
        'Country group': 'Country group'
    },
    template='plotly_white'
)

# Make the selected countries easier to identify
fig3.update_traces(marker=dict(size=7), selector=dict(name='Other countries'))
fig3.update_traces(marker=dict(size=13), selector=dict(name='Selected countries'))

# Add rank labels for Mexico and Spain
annotation_positions = {
    country1: {'ax': 35, 'ay': -35},
    country2: {'ax': 35, 'ay': 35}
}

for _, row in ranking_df[ranking_df['Country Code'].isin(selected_countries)].iterrows():
    position = annotation_positions.get(row['Country Code'], {'ax': 35, 'ay': -35})
    fig3.add_annotation(
        x=row['Rank'],
        y=row[ranking_year],
        text=f"{row['Country Code']} rank {int(row['Rank'])}",
        showarrow=True,
        arrowhead=2,
        ax=position['ax'],
        ay=position['ay']
    )

fig3.update_yaxes(tickprefix='$', tickformat='.2s')
fig3.update_layout(height=600, legend_title_text='Country group')

fig3.show()
```

## Pie Chart. Top 5 countries by GDP

Requirement covered: create a pie chart with the top five GDP countries in a year different from 2014 and 1975.

```python
# Requirement covered: pie chart for the top five countries by GDP
pie_df = df_new[['Country Name', 'Country Code', pie_year]].dropna().copy()
pie_df = pie_df.sort_values(by=pie_year, ascending=False).head(5).reset_index(drop=True)

fig4 = px.pie(
    pie_df,
    values=pie_year,
    names='Country Name',
    title=f'Pie Chart. Top 5 Countries by GDP in {pie_year}',
    template='plotly_white'
)

fig4.update_traces(
    textposition='inside',
    textinfo='percent+label',
    hovertemplate='Country: %{label}<br>GDP: $%{value:,.0f}<br>Share: %{percent}<extra></extra>'
)
fig4.update_layout(
    legend_title_text='Country',
    height=600
)

fig4.show()
```
